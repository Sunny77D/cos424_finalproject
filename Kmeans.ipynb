```python
import numpy as np
import pandas as pd
import sklearn as sklearn
from matplotlib import pyplot as plt
from PIL import Image
from sklearn.cluster import KMeans

```

```python
X = pd.read_csv('combined_data.csv',low_memory=False, header = None)
X = X.apply(pd.to_numeric, errors='coerce')
X = X.dropna() 
artist = X[4800]
X = X.drop(columns=4800)
X = np.asarray(X)
```

```python
#kmeans = KMeans(n_clusters=20).fit(X)
```

```python
x1 = range(1,26)
y1 = []
for i in range(25):
    print(i)
    kmeans = KMeans(n_clusters=i+1).fit(X)
    y1.append(kmeans.inertia_)

plt.plot(x1, y1)
plt.xlabel("Number of Clusters")
plt.ylabel("KMeans Objective Function Sum")
plt.title("KMeans Objective Function with vary Ks (Color)")
plt.show()
```

```python
kmeans = KMeans(n_clusters=10).fit(X)
fig=plt.figure(figsize=(6,6))
j = 0
for i in kmeans.cluster_centers_: 
    i = i.reshape((40,40,3))
    pic = i.astype(int)
    fig.add_subplot(5, 2, j+1)
    j += 1
    plt.imshow(pic)
plt.show()
```
