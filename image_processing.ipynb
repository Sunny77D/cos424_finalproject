```python
import numpy as np
from matplotlib import pyplot as plt
import pandas as pd
from matplotlib.image import imread
import PIL
from PIL import Image
import glob

```

```python
def rgb2gray(rgb):
    return np.dot(rgb[...,:3], [0.2989, 0.5870, 0.1140])
```

```python
# Sytle Transfer
test_img = Image.open('lewis_center.png')
res_img2 = test_img.resize((40, 40), Image.NEAREST)
plt.imshow(res_img2)
plt.show()

test_img = test_img.convert('RGB')
res_img = test_img.resize((40, 40), Image.NEAREST)
res_img = np.asarray(res_img)
#res_img = rgb2gray(res_img)
np_img = np.array(res_img)
fin_img = np_img.reshape((1,4800))
# Saving the vector
#np.savetxt("styletransfer.csv", fin_img, delimiter=",")
# Making sure you can get the image back
img_cov = np.reshape(fin_img, (40,40,3))
plt.imshow(img_cov)
plt.show()
```

```python
images = []
i = 0
for img in glob.glob("Paul_Klee/*.jpg"):
    n = Image.open(img)
    res_img = n.resize((40, 40), Image.NEAREST)
    np_img = np.array(res_img)
    if(np_img.shape != (40,40,3)):
        np_img2 = np_img.copy()
        np_img3 = np_img.copy()
        np_rbg = np.stack((np_img, np_img2, np_img3), axis=2)            
        fin_img = np_rbg.reshape((1,4800))
    else:
        fin_img = np_img.reshape((1,4800))
    b = np.array([37])
    b = b.reshape((1,1))
    fin_row = np.append(fin_img, b, axis=1)
    images.append(fin_row)
    i += 1

images = np.asarray(images) 
images = images.reshape((i, 4801))
np.savetxt("Paul_Klee.csv", images, delimiter=",")
```

```python
images = []
i = 0
for img in glob.glob("William_Turner/*.jpg"):
    n = Image.open(img)
    res_img = n.resize((40, 40), Image.NEAREST)
    np_img = np.asarray(res_img)
    if (np_img.shape != (40, 40)):
        np_img = rgb2gray(np_img)    
    fin_img = np_img.reshape((1,1600))
    b = np.array([50])
    b = b.reshape((1,1))
    fin_row = np.append(fin_img, b, axis=1)
    images.append(fin_row)
    i += 1

images = np.asarray(images) 
images = images.reshape((i, 1601))
np.savetxt("William_Turner-gray.csv", images, delimiter=",")
```
